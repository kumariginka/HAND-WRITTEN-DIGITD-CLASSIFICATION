```python
####scaaling doen of the pixels 
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
train_images[0]
```

##### reshaping of our parameter to 1d  OR  we can use keras Flatten  layer

```python
train_images=train_images.reshape(len(train_images),28*28)
test_images=test_images.reshape(len(test_images),28*28)
```

```python
model=keras.Sequential([
    #keras.layers.Flatten(input_shape=(28,28)),            ###flatting  our inputs
    keras.layers.Dense(units=10,
                       activation="sigmoid",
                       input_shape=(784,))
                       
                      ]  )                   
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
model.fit(train_images,train_labels,epochs=10)
```

```python
model.evaluate(test_images,test_labels)
```

```python
yp=model.predict(test_images)
```

```python
yp
```

```python
yp_predicted=[np.argmax(i) for i in yp]
```

```python
yp_predicted
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(test_labels,yp_predicted)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(cm,annot=True)
```

# i want to improve accuracy by adding hidden layers

# instead of doing manual flattening i am gona use of keras flatten layer

#### i am gona add more hidden layers to keep accuracy in track

```python
model=keras.Sequential([
    keras.layers.Flatten(),        ###### here i was again reload data im not done manuaaly flattening ...i was just flatted here use of this layer 
    keras.layers.Dense(units=100,activation="relu"),   ####hidden layer 1
    keras.layers.Dense(units=80,activation="relu"), 
    keras.layers.Dense(units=70,activation="relu"),###hidden layer 2
    keras.layers.Dense(units=60,activation="relu"),       #####hidden layer 3
    keras.layers.Dense(units=40,activation="relu"), 
    keras.layers.Dense(units=30,activation="relu"),##hidden layer 4
    keras.layers.Dense(units=20,activation="relu"),         ###hidden layer 5
    keras.layers.Dense(units=10,activation="sigmoid")
    
])
model.compile(optimizer="adam",
             loss="sparse_categorical_crossentropy",
             metrics=["accuracy"])
model.fit(train_images,train_labels,epochs=10)
```

```python
model.evaluate(test_images,test_labels)
```

```python
kumari=model.predict(test_images)
```

```python
kumari     #####results are in the form of array 
```

```python
### to convert array to desired labels
predicted_results=[np.argmax(i) for i in kumari]
```

```python
cm=confusion_matrix(test_labels,predicted_results)
```

```python
plt.figure(figsize=(12,10))                ######  off diagonal elements values are decreases
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(test_labels,predicted_results))
```
